# Review of Sorting Alogorithms

**A sorting algorithm** arranges the elements in a list or collection in increasing or decreasing order of some property

Characteristics that define sorting algorithms:
1. Time complexity
2. Space Complexity / Memory Storage
    - in-place, constant memory, or memory usage that grows with input size?
3. Stability 
4. Internal sort versus external sort
    - all records in main memory, or external memory?
5. Recursive or non-recursive

## O(N) time complexity sorting algorithms

**Selection Sort**
- find (select), the minimum (or maximum), and move it to the start of another array or list
- continue the steps until the first list is empty

**Bubble Sort**
- scan the array from left to right, if the item at index i is greater than item at index i+1, we swap the elements
- when we hit the end of the list, we restart
- we repeat the process until a pass with no swaps are made

**Insertion Sort**
- break array / list into two subsections, sorted and non-sorted
- we iterate through the unsorted items, for each element in the sorted section that is greater than the element, we move one to the right, then we insert the element into the appropriate index of sorted
- we repeat the process until the unsorted section is empty (in reality we are constantly increasing the boundry of the sorted)




```python
#Selection Sort
a = [7,2,4,1,5,3]

def selection_sort(array):
    
    a2 = []
    for i in range(0,len(array)):
        a2.append(min(array))
        array.remove(min(array))
    print(a2)
    #print(a)
```

```python
#Bubble Sort
a = [7,2,4,1,5,3]

def bubble_sort(array):
    
    #iterate through list
    for k in range(0,len(array)-1):
        swaps = False
        #peform the scan
        for i in range(0,len(array)-k-1):
            if array[i] > array[i+1]:
                #swap
                temp = array[i+1]
                array[i+1] = array[i]
                array[i] = temp
                swaps = True
        if swaps == False:
            return array
    
```

```python
#Insetion Sort
a = [7,2,4,1,5,3]

def insertion_sort(array):
    
    for i in range(1,len(array)-1):
        value = array[i]
        hole = i
        while (hole > 0 and array[hole-1]>value):
            array[hole] = array[hole-1]
            hole -= 1
        array[hole] = value
    return array
    

                
    
    

# - break array / list into two subsections, sorted and non-sorted
# - we iterate through the unsorted items, for each element in the sorted section that is greater than the element, we move one to the right, then we insert the element into the appropriate index of sorted
# - we repeat the process until the unsorted section is empty (in reality we are constantly increasing the boundry of the sorted)
```

## NlogN time complexity sorting algorithms

**Merge Sort**
- A recursive algorithm
- given an array, and two sorted halves of that array, scan both halved arrays, 'merging' the minimum of the two elements into the array
- if one of the halfed array becomes empty, than we simply merge the rest of the other array
- in order to obtain the two soted halves, we recursive perform this algorithm, until we are just comparing two elements (divide and conquer)



**Quick Sort**
- A recursive algorithm
- Given a partitioned array, where all elements left of the pivot are lesser, and all elements right of the pivot are equal or greater
- we iteratively break the array in two and perform a partition, until there is only one item to partition, At that point, we know the elements are sorted




```python
#mege sort
a = [7,2,4,1,5,3]

#Algorithm that will merge to sorted arrays into the original array (the combine step):
def merge(L,R,A):
    i = 0
    j = 0
    k = 0
    
    while k < len(A):
        while (i < len(L) and j < len(R)):
            if L[i] < R[j]:
                A[k] = L[i]
                i += 1
            else:
                A[k] = R[j]
                j += 1
            k += 1
        while j < len(R):
            A[k] = R[j]
            j += 1
            k += 1
        while i < len(L):
            A[k] = L[i]
            i += 1
            k += 1
    return A

#Algorithm that will recursively divide into smaller arrays, then combine them
def merge_sort(A):
    n = len(A) #6 -> 3
    if n < 2:  
        return
    else:
        mid = int(n/2) #3 -> 
        L = A[0:mid] #0:3
        R = A[mid:n] #3:6
        #recursively calling function
        merge_sort(L) #[7,2,4]
        merge_sort(R) #[1,5,3]
        
        merge(L,R,A)
        
    return A
```
